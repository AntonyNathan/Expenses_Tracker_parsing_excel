```python
import csv
```

```python
data = open('ExpensesSheet2.csv',encoding="utf-8")
```

```python
csv_data = csv.reader(data)
data_lines = list(csv_data)
```

```python
data_lines[2][1]
```

```python
Groceries_list = []
food_list = []
Salary_list = []
Dentist_list = []
Transport_list = []
Rent_list = []
```

```python
for i in range(0,162):
    if(data_lines[i][0]=='Groceries'):
            Groceries_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(Groceries_list)):
        print(Groceries_list[x])
        
for i in range(0,162):
    if(data_lines[i][0]=='Food'):
            food_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(food_list)):
        print(food_list[x])

for i in range(0,162):
    if(data_lines[i][0]=='Transport'):
            Transport_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(Transport_list)):
        print(Transport_list[x])

for i in range(0,162):
    if(data_lines[i][0]=='Salary'):
            Salary_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(Salary_list)):
        print(Salary_list[x])
    
for i in range(0,162):
    if(data_lines[i][0]=='Dentist'):
            Dentist_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(Dentist_list)):
        print(Dentist_list[x])
        
for i in range(0,162):
    if(data_lines[i][0]=='Rent'):
            Rent_list.append(float(data_lines[i][1]))
    else:
        continue
    for x in range(0,len(Rent_list)):
        print(Rent_list[x])
```

```python
Foodprice = [float(x) for x in food_list]
Salaryamount = [float(x) for x in Salary_list]
Groceriesprice = [float(x) for x in Groceries_list]
Transportprice = [float(x) for x in Transport_list]
Dentistprice = [float(x) for x in Dentist_list]
Rentprice = [float(x) for x in Rent_list]
```

```python
Tot_food = sum(Foodprice)
Tot_salary = sum(Salaryamount)
Tot_groceries = sum(Groceriesprice)
Tot_transport = sum(Transportprice)
Tot_dentist = sum(Dentistprice)
Tot_rent = sum(Rentprice)
```

```python
print ('Amount spent on food:', Tot_food)
print ('Amount spent on Groceries:', Tot_groceries)
print ('Amount spent on Transport:', Tot_transport)
print ('Amount spent on dentist:', Tot_dentist)
print ('Amount spent on rent:', Tot_rent)
print ('Amount earned on Salary:', Tot_salary)
```
